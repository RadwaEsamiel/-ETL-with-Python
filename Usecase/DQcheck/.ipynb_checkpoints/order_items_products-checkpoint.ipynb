```python
import sqlalchemy as db
import pandas as pd
import numpy as np


path = "/home/radwa/Radwa.Usecase/Landing/order_items/order_items.csv"

df_order_items = pd.read_csv(path)
```

```python
df_order_items.isnull().sum()
```

```python
df_order_items.info()
```

```python
duplicate_ids = df_order_items['order_id'].duplicated() 

df_duplicate_ids = df_order_items[duplicate_ids]

print("Rows with duplicate 'id':")
print(df_duplicate_ids)
```

```python
duplicate_rows = df_order_items.duplicated()  

df_duplicates = df_order_items[duplicate_rows]

print("Duplicate rows:")
print(df_duplicates)
```

```python
df_order_items.drop_duplicates(inplace=True, keep="first")
```

```python
df_order_items["item_id"] = pd.to_numeric(df_order_items["item_id"], errors='coerce')  
```

```python
df_order_items["item_id"].fillna(0, inplace=True)
```

```python
df_order_items["item_id"] = df_order_items["item_id"].astype(int)
```

```python
start_price = 200
end_price = 10000  

is_within_range = df_order_items["list_price"].between(start_price, end_price)

df_order_item_table = df_order_items[is_within_range] 
```

```python
df_order_item_table
```

```python

path = "/home/radwa/Radwa.Usecase/Landing/Products/products.csv"

df_products = pd.read_csv(path)
```

```python
distinct_item_ids = df_order_items['item_id'].unique()
```

```python
df_products_filtered = df_products[df_products['product_id'].isin(distinct_item_ids)]
```

```python
df_products_filtered
```

```python
df_products_filtered.isnull().sum()
```

```python
import os

csv_path = "/home/radwa/Radwa.Usecase/staging_1/Products/products.csv" 

directory = os.path.dirname(csv_path)
if not os.path.exists(directory):
    os.makedirs(directory)
    
df_products_filtered.to_csv(csv_path, index=False)
```
